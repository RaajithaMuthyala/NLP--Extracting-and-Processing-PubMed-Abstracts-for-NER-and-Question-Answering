```python
import spacy
import scispacy
from scispacy.linking import EntityLinker
import pandas as pd
```

```python
ner = spacy.load("en_core_sci_scibert")
```

```python
data = pd.read_csv('/content/drive/MyDrive/Raajitha_NLP_CourseCapstone/processed_pubmed_abstracts.csv')
```

```python
data.head()
```

```python
%time
ner.add_pipe("scispacy_linker", config={"resolve_abbreviations": True, "linker_name": "umls"})
```

```python
# Function to extract UMLS information
def extract_umls_entities(text):
    doc = ner(text)  # Process the text with the NER model
    entity_details = []  # List to store entity details

    # Iterate over entities and extract UMLS information
    for entity in doc.ents:
        if entity._.kb_ents:  # Check if entity is linked to UMLS
            cui, score = entity._.kb_ents[0]  # Get the top UMLS concept
            kb_entry = ner.get_pipe("scispacy_linker").kb.cui_to_entity[cui]
            # Append extracted information
            entity_details.append({
                "CUI": cui,
                "Name": kb_entry.canonical_name,
                "Definition": kb_entry.definition,
                "TUIs": kb_entry.types,
                "Aliases": kb_entry.aliases[:5]  # Top 5 aliases
            })

    return entity_details

# Apply the function to the 'Abstract' column
data['umls_entities'] = data['tokenized_abstract'].apply(extract_umls_entities)
```

```python
# Flatten the extracted entity details into rows for each entity
entity_rows = []
for idx, row in data.iterrows():
    for entity in row['umls_entities']:
        entity_row = {
            "Abstract_ID": idx,  # Add Abstract ID to identify the source
            **entity  # Unpack entity information
        }
        entity_rows.append(entity_row)

# Create a new DataFrame for the entities
entity_df = pd.DataFrame(entity_rows)

# Display the first few rows to verify the structure
print(entity_df.head())
```

```python
# Save to CSV
entity_df.to_csv('/content/drive/MyDrive/Raajitha_NLP_CourseCapstone/extracted_umls_entities.csv', index=False)
```
