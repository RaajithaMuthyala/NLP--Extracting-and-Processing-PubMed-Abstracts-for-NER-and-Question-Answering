```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Raajitha_NLP_CourseCapstone/extracted_umls_entities.csv')
```

```python
data.head()
```

```python
# View unique entities and their counts
unique_entities = data['Name'].value_counts()
print(unique_entities.head(10))
```

```python
# View unique CUIs
unique_cuis = data['CUI'].nunique()
print(f"Number of unique CUIs: {unique_cuis}")
```

```python
# View TUI (Type Unique Identifier) counts
tui_counts = data.explode('TUIs')['TUIs'].value_counts()
print(tui_counts)
```

```python
grouped_entities = data.groupby('Abstract_ID')['Name'].apply(list).reset_index()

# Rename columns for clarity
grouped_entities.columns = ['Abstract_ID', 'Entities']
```

```python
grouped_entities.to_csv('/content/drive/MyDrive/Raajitha_NLP_CourseCapstone/grouped_entities_by_abstract.csv', index=False)
```
